```python
dep = df[['Mileage', 'Age(yrs)']]
sp = df[['Sell Price($)']]

sp
```

```python
from sklearn.model_selection import train_test_split
```

```python
dep_train,dep_test, sp_train, sp_test = train_test_split(dep, sp, test_size = 0.2,random_state=10)
```
