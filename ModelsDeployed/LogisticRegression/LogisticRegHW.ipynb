```python
subdf = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
subdf.head()
```

```python
salary_dummies = pd.get_dummies(df.salary, prefix = "salary_")
```

```python
salary_dummies
```

```python
df = df.drop(['salary'], axis = 'columns')
```

```python
df
```

```python
subdf = pd.concat([subdf, salary_dummies], axis = 'columns')
```

```python
subdf
```

```python
left2 = df[['left']]
left2
```

```python
subdf =subdf.drop('salary', axis = 'columns')
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(subdf, left2 , train_size = 0.7)
```

```python
lg.fit(X_train, Y_train)
```

```python
lg.score(X_test, Y_test)
```
