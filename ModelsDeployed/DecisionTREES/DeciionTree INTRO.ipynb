```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
```

```python
target = df['salary_more_then_100k']
target


```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
company_le = le.fit_transform(df['company'])
job_le = le.fit_transform(df['job'])
degree_le = le.fit_transform(df['degree'])

df['company_le'] = company_le
df['job_le'] = job_le
df['degree_le'] = degree_le
df
```

```python
inputs = df.drop(['salary_more_then_100k','degree','company', 'job'], axis= 'columns')
target = df['salary_more_then_100k']
target
inputs
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs, target)
```

```python
model.score(inputs, target)
```
