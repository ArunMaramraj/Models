```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv")
df.head(10)
```

```python
#survived , pclass , sex, age, fair -- where survived is the dependent variable
inputs = df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']]
inputs
```

```python
newinputs = df.drop(columns = ['Survived'])
target = df['Survived']
target
```

```python
mean = inputs['Age'].mean()
mean
        
```

```python
inputs['Age']= inputs['Age'].fillna(mean)
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
inputs['Sex_le'] = le.fit_transform(inputs['Sex'])

newinputs = inputs.drop(columns = ['Survived','Sex','Sex-le'])

newinputs
```

```python
newinputs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(newinputs,target,test_size=0.5) 
```

```python
from sklearn import tree

```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test, Y_test)
```
